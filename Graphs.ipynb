# 3. Histogram

```python
# bar plot :- For discrete
# histogram :- For continues data
# in the sense that a bar graph relates two variables, but a histogram relates only one
```

```python
'''Histogram Plot:- we use histogram it define ranges(bin) then count or define frequncy(0 to 100) me kitne order huee h
                    bins:- How many parts we devide it
1. For continues data
2. In the sense that a bar graph relates two variables, but a histogram relates only one
3. Histogram plot for continues value that's wht it has not gape if there is gape then it's mean
   there is not any value
4. For countplot of continues data we use histogram it define ranges(bin) then count or define frequncy

Links:- 
1. https://statistics.laerd.com/statistical-guides/understanding-histograms.php
2. https://www.youtube.com/watch?v=snkkKrek7TU
3. https://www.statisticshowto.datasciencecentral.com/choose-bin-sizes-statistics/
'''
```

```python
# bin = take that nu which can perfectley devide the (min value of data - max value of data)
# bin = 1. whole number
#       2. (min value of data - max value of data) % n == 0
#       3. bin size = (min value of data - max value of data) / bin
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('num_orders')
plt.ylabel('frequency')
ax = plt.hist(df['num_orders']) 
```
